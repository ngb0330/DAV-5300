# Research Question

This project will look at the relationship between student SAT scores on the verbal and math sections. <br>
The SAT is a standardized test widely used for college admissions in the United States. <br>
The data for this project is taken from [statcrunch.com](https://www.statcrunch.com/app/index.php?dataid=376459). <br>
The dataset contains 162 rows, and contains three columns: Verbal (the score on the verbal section of the SAT), Math (the score on the math section of the SAT), and Sex (the sex of the test taker). <br>

The question that I am seeking to answer is:
>"Is there a relationship between SAT scores on the verbal section and the math section or sex?"

My null hypothesis is: There is no relationship between SAT scores on the verbal section and SAT scores on the math section or sex.

My alternative hypotheses are: 
1. There is a relationship between SAT scores on the verbal section and SAT scores on the math section. 
2. There is a relationship between SAT scores on the verbal section and sex.

# Tables of Regression Results

In order to examine the relationship between scores on the verbal section and the math section, I created a linear regression using the statsmodels library between the two variables. <br>
The linear regression showed that there was a positive linear relationship between scores on the math section and on the verbal section. <br>
For each additional point scored on the math section, 0.6943 additional points would be expected to be scored on the verbal section. <br>

The model has an adjusted R-squared of 0.465 which means that 46.5% of the variability in the data can be explained by the model. <br>
The Math variable has a p-value of 0.000, indicating that it is significant at a 95% confidence level. <br>


```python
import statsmodels.api as sm
# create a least squares regression of Math vs. Verbal
model = sm.OLS(sat['Verbal'], sm.add_constant(sat['Math']))
# fit the model
results = model.fit()
# print the results of the regression
print(results.summary())
```

In order to examine the relationship between scores on the verbal section and the sex of the test taker, I created a linear regression using the statsmodels library between the the Verbal and Is_Female columns of the sat DataFrame. <br>
The linear regression showed that there was a positive linear relationship between scores on the verbal section and being female. <br>
If the test taker is female, we would expect the score on the verbal section to be 11.6982 points higher. <br>

The model has an adjusted R-squared of -0.003 which means that none of the variability in the data can be explained by the model. <br>
The Is_Female variable has a p-value of 0.456, indicating that it is not significant at a 95% confidence level.

```python
# create a least squares regression of Is_Female vs. Verbal
model = sm.OLS(sat['Verbal'], sm.add_constant(sat['Is_Female']))
# fit the model
results = model.fit()
# print the results of the regression
print(results.summary())
```

I created a multiple linear regression to show the relationship between both the math section and the gender of the test taker and the verbal section of the SAT.<br>
The linear regression showed that there was a positive linear relationship between scores on the math section and on the verbal section, and between being female and scores on the verbal section.  <br>

All other things being equal, for each additional point scored on the math section, 0.7192 additional points would be expected to be scored on the verbal section. <br>
All other things being equal, if the test taker is female, we would expect the score on the verbal section to be 32.930 points higher. <br>

The model has an adjusted R-squared of -0.489 which means that 48.9% of the variability in the data can be explained by the model. <br>
The Math variable has a p-value of 0.000, indicating that is significant at a 95% confidence level.
The Is_Female variable has a p-value of 0.004, indicating that it is significant at a 95% confidence level.

```python
import statsmodels.formula.api as smf
# create a least squares regression of Math and Is_Female vs. Verbal
model = smf.ols(formula = 'Verbal ~ Math + Is_Female', data = sat)
# fit the model
result = model.fit()
# print the results of the regression
print(result.summary())
```

# Final Conclusions

The first linear regression showed the relationship between the math section and the verbal section indicated that for each additional point scored on the math section, we would expect to score an additional 0.6943 points on the verbal section. <br>
This relationship seemed to be significant as it had a p-value of 0.000. <br>
At a 95% confidence level, the expected increase in the score on the verbal section is between 0.579 and 0.810 for each additional point on the math section. <br>

The second linear regression showed the relationship between the sex of the test taker and the verbal section indicated that if the test taker was female, we would expect the score on the verbal section to be 11.6982 points higher. <br>
This relationship seemed to be non-significant as it had a p-value of 0.456. <br>
At a 95% confidence level, the expected increase in the score on the verbal section is between -19.230 and 42.627 if the test taker was female.<br>
This interval includes both negative numbers, which means that it is possible that being female has a negative impact on scores on the verbal section, and zero, which would indicate that being female has no impact on the score on the verbal section.<br>

The final linear regression showed the relationship between the math section and sex of the test taker and the verbal section. <br>
It indicated that all other things being equal, for each additional point scored on the math section, 0.7192 additional points would be expected to be scored on the verbal section. and that all other things being equal, if the test taker is female, we would expect the score on the verbal section to be 32.930 points higher. <br>>
This relationship seemed to be significant as the Math variable had a p-value of 0.000, and the Is_Female variable had a p-value of 0.004. <br>
At a 95% confidence level, the expected increase in the score on the verbal section is between 0.605 and 0.833 for each additional point on the math section. <br>
At a 95% confidence level, the expected increase in the score on the verbal section is between 10.361 and 55.025 if the test taker was female. <br>

Based on the results of this analysis we reject the null hypothesis that there is no relationship between scores on the verbal section of the SAT, and scores on the math section or sex. <br>
This is because the p-values of the of Math and Is_Female variables in the multiple regression make it extremely unlikely that the positive linear relationship observed in the model are due to random chance. <br>
The significance level was set at 95%, and both p-values were within that threshold. <br>

However, on its own, the sex of the test taker does not influence the score on the verbal section, as in the single variable linear regression, the p-value of the Is_Female variable was greater than 0.05, and the confidence interval included zero. <br>
On its own, the score on the math section does appear to still influence the score on the verbal section, as in the single variable linear regression, the p-value of the Math variable was less than 0.05, and the confidence interval did not include zero. <br>
